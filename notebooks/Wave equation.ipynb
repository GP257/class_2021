Lets begin by writing the elastic wave equation in terms of particle velocities ($V$), stresses ($\sigma$), density ($\rho$), the Lame parameters ($\lambda$ and $\mu$), and sources $s$.

\label{elasticeqn3}

To turn partial differential equations into a finite difference scheme lets do a brief review of the Taylor series. Normally we write the Taylor series of $\Delta x$, here we will write in terms of successive grid points. Next we write it in terms of $x_{i_1}$, $x_i$, amd $x_{i+1}$.

\begin{eqnarray} 
f(x_{i-1}) & =& f(x_i)+(x_{i+1}-x_i)f'(x_i)+ \frac{(x_i+1-x_i)^2}{2!} f''(x_i) - \frac {(x_{i+1}-x_i)^3}{3!} f''(x_i)+.... \\
f(x_i) &=& f(x_i) \\
f(x_{i+1}) & =& f(x_i)+(x_{i+1}-x_i)f'(x_i)+ \frac{(x_i+1-x_i)^2}{2!} f''(x_i) + \frac {(x_{i+1}-x_i)^3}{3!} f''(x_i)+....
\end{eqnarray}

Which when we take into account are sampling becomes
\begin{eqnarray} 
f(x_{i-1}) & =& f(x_i)-\Delta x f'(x_i)+ \frac{\Delta x^2}{2!} f''(x_i) - \frac {\Delta x ^3}{3!} f''(x_i)+.... \\
f(x_i) &=& f(x_i) \\
f(x_{i+1}) & =& f(x_i)+\Delta x f'(x_i)+ \frac{\Delta x^2}{2!} f''(x_i) + \frac {\Delta x^3}{3!} f''(x_i)+....
\end{eqnarray}

We will aproximate our first order derivative subtracting the third equation from the second and ignoring but the first two terms.
\begin{equation}
f(x_{i+1})- f(x) = f(x_i) + \Delta x f'(x_i)-f(x_i),
\end{equation}

which simplifies to 
\begin{equation}
f'(x_i) = \frac {f(x_{i+1}) -f(x_i)} {\Delta x} .
\end{equation}

The terms that we truncated represent the error in our approximation. So for this case we have an error 
of
\begin{equation}
E=\frac{\Delta x^2}{2!} f''(x_i) + \frac {\Delta x ^3}{3!} f''(x_i)+.... 
\end{equation}

Note if we subtract the first from the third equation we get the central difference
\begin{equation}
f'(x_i) = \frac {f(x_{i+1}) -f(x_i-1)} {2 \Delta x} 
\end{equation}
with an error of
\begin{equation}
E=\frac{2 (\Delta x)^2} {3!} f''(x_i)
\end{equation}

 The error causes what is called numerical dispersion. In the figure below notice how we have what looks like ringing, different frequencies are moving at difference speeds. 

<img src="image.png" alt="Drawing" style="width: 200px;"/>
We can reduce this dispersion by more finely sampling finite difference approximation or increasing the accuracy of our derivative approximation. For our time derivatives in our wave equation we will have to pay the cost of smaller time steps. For spatial derivative we will try to further improve the accuracy of derivative approximation.

Lets go back to our Taylor expansion add some additional point away from the location where we want to calcuate our derivative, and some further terms.
\begin{eqnarray} 
f(x_{i-2}) & =& f(x_i)- 2 \Delta x f'(x_i)+ \frac{(2 \Delta x)^2}{2!} f''(x_i) - \frac {(2 \Delta x)^3}{3!} +\frac{(2 \Delta x)^4}{4!} f''''(x_i)- \frac{(2 \Delta x)^5}{5!} f'''''(x_i)+ \\
f(x_{i-1}) & =& f(x_i)-\Delta x f'(x_i)+ \frac{\Delta x^2}{2!} f''(x_i) - \frac {\Delta x ^3}{3!} f''(x_i)+\frac{( \Delta x)^4}{4!} f''''(x_i)- \frac{( \Delta x)^5}{5!} f'''''(x_i) \\
f(x_i) &=& f(x_i) \\
f(x_{i+1}) & =& f(x_i)+\Delta x f'(x_i)+ \frac{\Delta x^2}{2!} f''(x_i) + \frac {\Delta x^3}{3!} f''(x_i)+\frac{( \Delta x)^4}{4!} f''''(x_i)+ \frac{(\Delta x)^5}{5!} f'''''(x_i) \\
f(x_{i+2}) & =& f(x_i)+ 2 \Delta x f'(x_i)+ \frac{(2 \Delta x)^2}{2!} f''(x_i)+\frac {(2 \Delta x)^3}{3!} f''(x_i)+ \frac{(2 \Delta x)^4}{4!} f''''(x_i)+ \frac{(2 \Delta x)^5}{5!} f'''''(x_i)  \\
\end{eqnarray}

What we have is an inverse problem in disguise.  What we to know is how to scale and sum the points $f(x_{i-2})$ ... $f(x_{i+2})$ to best approximate a first derivative
\begin{equation}
f'(x_i) = A f(x_{i-2}) +B f(x_{i-1}) + C f(x_i) + D f(x_{i+1}) + E f (x_{i+2})
\end{equation}

Which means we want to cancel out all of the higher derviates and the non derivative in the above set of equations. So what we get is a simple matrix inverse probkem. For the above example our problem looks like

\begin{equation}
\begin{bmatrix}
           0 \\
           d! \\
           0 \\
           0 \\
           0 \\
         \end{bmatrix} = 
\left[
  \begin{array}{ccc}
    1 & 1 & 1 & 1 & 1\\
    -2 &-1& 0 & 1 & 2 \\
    4  & 1& 0 & 1 & 4 \\
    -8 & -1 & 0 & 1 & 8 \\
    16 & 1 & 0 & 1 &16 \\
    -32 & -1 & 0 & 1 &32 \\
  \end{array}
\right]
\begin{bmatrix}
          A \\
          B \\
           C \\
           D \\
           E \\
         \end{bmatrix} ,
\end{equation}
where $d$d is the derivative we want to calculate.
We can generalize how to calculate coeficents with the function below.

```python
import math
import numpy as np
def calcDerivCoefs(offsets,derivative):
    array=np.ndarray((len(offsets),len(offsets)))
    data=np.ndarray(len(offsets))
    array[0,:]=1
    data[0]=0
    tm=1
    for ideriv in range(1,len(offsets)):
        tm=tm*ideriv
        if derivative==ideriv:
            data[ideriv]=tm
        else:
            data[ideriv]=0
        for icoef in range(len(offsets)):
            array[ideriv,icoef]=math.pow(offsets[icoef],ideriv)
    mod=np.linalg.solve(array,data)
    return mod
```

```python
import numba
import numpy as np
class earthP:
    def __init__(self,ns,ds,val=None):
        self._n=ns
        self._d=ds
        self._vals=np.ndarray(ns)
        if val != None:
            self._vals[:,:,:]=val

            
class extractFunc:
    def __init__(self):
        pass;
    def extractData(self,wfld,it):
        raise Exception("Must override extractData")

class extractSlice1(extractFunc):
    def __init__(self,islice,wfld,nt,jt):
        self._vals=np.ndarray((math.ceil(nt/jt),wfld._n[1],wfld._n[2]),np.float32)
        self._is=islice
        self._jt=jt
    def extractData(self,wfld,it):
        if it%self._jt==0:
            i=int(it/self._jt)
            self._vals[i,:,:]=wfld[self._is,:,:]
        
class sourceFunc:
    def __init__(self,maxF):
        self._maxF=maxF
    def injectSource(self,field,t):
        """Inject source into a field"""
        raise Exception("Must override injectSource")

class ricker(sourceFunc):
    def __init__(self,freq,dt,length,x,y,z,model,d):
        t = np.arange(-length/2, (length-dt)/2, dt)
        self._vals = (1.0 - 2.0*(np.pi**2)*(freq**2)*(t**2)) * np.exp(-(np.pi**2)*(freq**2)*(t**2))
        
        self._dt=dt
        self._ix=int(x/d[0])
        self._iy=int(x/d[1])
        self._iz=int(x/d[2])
        self._first=0
        if self._ix < 0 or self._ix >= model.shape[0] or self._iy<0 or self._iy >model.shape[1] or self._iz <0 or self._iz >= model.shape[2]:
            raise Exeption("Source is outside model")
        super().__init__(freq*2.)
    def injectSource(self,field,t):
        injectNumba(field,t,self._ix,self._iy,self._iz,self._dt,self._vals)



@numba.njit()
def injectNumba(field,t,ix,iy,iz,dt,wave):
    ft=t/dt
    if ft < wave.shape[0]:
        it=int(ft)
        ft-=it
    field[ix,iy,iz]=wave[it]*(1.-ft)+wave[it+1]*ft
```

And now our first version of a class that does acoustic propagation with density. Note how we have to be careful with our boundary conditions, the loops on calcuating the derivatives for velocities and stresses.

```python
import math
class acousticDensity:
    def __init__(self,density,lamda,source,extract,dt):
        """
        density - Density as a function of space
        lamda - Lamda as a function of space
        nt    - Number of time samples
        source - Source function
        extract - Extract from wavefield
        
        """
        if not isinstance(density,earthP):
            raise Exception("Expecting density to be of type earthP")
        if not isinstance(lamda,earthP):
            raise Exception("Expecting lamda|to be of type earthP")
        if not isinstance(source,sourceFunc):
            raise Exception("Expecting source to be of type sourceFunc")
        if not isinstance(extract,extractFunc):
            raise Exception("Expecting extract to be of type extractFunc")
        self._extract=extract
        self._source=source
        self._lamda=lamda
        self._invD=density
        self.createVectors()

        minV=math.sqrt(lamda._vals[0,0,0]/density._vals[0,0,0])
        maxV=minV
        self._dt=dt
        dmin=min(min(lamda._d[0],lamda._d[1]),lamda._d[2])
        dmax=max(max(lamda._d[0],lamda._d[1]),lamda._d[2])

        for i3 in range(lamda._n[0]):
            for i2 in range(lamda._n[1]):
                for i1 in range(lamda._n[2]):
                    v=math.sqrt(lamda._vals[i3,i2,i1]/density._vals[i3,i2,i1])
                    minV=min(minV,v)
                    maxV=max(maxV,v)
                    self._invDN._vals[i3,i2,i1]=1./self._invDN._vals[i3,i2,i1]
        print("what",maxV,dt,dmin,source._maxF,dmax)
        print(maxV*dt/dmin,"=stability disperseive=",minV/source._maxF/dmax)
        if maxV*dt/dmin>=.5:
            raise Exception("Unstable")
        if minV/source._maxF/dmax <=2.8:
            raise Exception("Dispersive")
        
        cs=calcDerivCoefs([-4.5,-3.5,-2.5,-1.5,-.5,.5,1.5,2.5,3.5,4.5],1)
        for i in range(5):
            self._x[i]=cs[i]/self._lamda._d[0]
            self._y[i]=cs[i]/self._lamda._d[1]
            self._z[i]=cs[i]/self._lamda._d[2]

    def createVectors(self):
        self._invDN=self._invD
        self._Vx=np.ndarray(self._lamda._n,np.float32)
        self._Vy=np.ndarray(self._lamda._n,np.float32)
        self._Vz=np.ndarray(self._lamda._n,np.float32)
        self._prev=np.ndarray(self._lamda._n,np.float32)
        self._cur=np.ndarray(self._lamda._n,np.float32)
        self._next=np.ndarray(self._lamda._n,np.float32)
        self._z=np.ndarray((5,))
        self._y=np.ndarray((5,))
        self._x=np.ndarray((5,))
        self._cur.fill(0.)
        self._next.fill(0.)
        self._prev.fill(0.)
        self._Vx.fill(0.)
        self._Vy.fill(0.)
        self._Vz.fill(0.)

        
    def flipP(self):
        n=self._prev
        self._prev=self._cur
        self._cur=self._next
        self._next=n
        n[:,:,:]=0

    def prop(self,nt):
        t=0
        for i in range(nt):
            self.stepField()
            t+=self._dt
            self._extract.extractData(self._next,i)
            self._source.injectSource(self._next,t)
            self.flipP()
    def stepField(self):
        oneStep(self._Vx,self._Vy,self._Vz,self._invD._vals,self._lamda._vals,self._cur,self._prev,self._next,self._dt,self._x,self._y,self._z)

    def oneStep(self,Vx,Vy,Vz,invD,lamda,cur,prev,nex,dt,x,y,z):
        for i3 in range(5,Vx.shape[0]-5):
            for i2 in range(5,Vx.shape[1]-5):
                for i1 in range(5,Vx.shape[2]-5):
                    Vx[i3,i2,i1]=lamda[i3,i2,i1]*(\
                    x[4]*(cur[i3,i2,i1-1]-cur[i3,i2,i1])+\
                    x[3]*(cur[i3,i2,i1-2]-cur[i3,i2,i1+1])+\
                    x[2]*(cur[i3,i2,i1-3]-cur[i3,i2,i1+2])+\
                    x[1]*(cur[i3,i2,i1-4]-cur[i3,i2,i1+3])+\
                    x[0]*(cur[i3,i2,i1-5]-cur[i3,i2,i1+4])\
                                                 )

                    Vy[i3,i2,i1]=lamda[i3,i2,i1]*(y[4]*(cur[i3,i2-1,i1]-cur[i3,i2,i1])+\
                    y[3]*(cur[i3,i2-2,i1]-cur[i3,i2+1,i1])+y[2]*(cur[i3,i2-3,i1]-cur[i3,i2+2,i1])+\
                    y[1]*(cur[i3,i2-4,i1]-cur[i3,i2+3,i1])+y[0]*(cur[i3,i2-5,i1]-cur[i3,i2+4,i1]))    
                    
                    Vz[i3,i2,i1]=lamda[i3,i2,i1]*(z[4]*(cur[i3-1,i2,i1]-cur[i3,i2,i1])+\
                    z[3]*(cur[i3-2,i2,i1]-cur[i3+1,i2,i1])+z[2]*(cur[i3-3,i2,i1]-cur[i3+2,i2,i1])+\
                    z[1]*(cur[i3-4,i2,i1]-cur[i3+3,i2,i1])+z[0]*(cur[i3-4,i2,i1]-cur[i3+4,i2,i1]))  


        dt2=dt*dt
        for i3 in range(10,nex.shape[0]-10):
            for i2 in range(10,nex.shape[1]-10):
                for i1 in range(10,nex.shape[2]-10):                   
                    a1=(x[4]*(Vx[i3,i2,i1]-Vx[i3,i2,i1+1])+\
                    x[3]*(Vx[i3,i2,i1-1]-Vx[i3,i2,i1+2])+x[2]*(Vx[i3,i2,i1-2]-Vx[i3,i2,i1+3])+\
                    x[1]*(Vx[i3,i2,i1-3]-Vx[i3,i2,i1+4])+x[0]*(Vx[i3,i2,i1-4]-Vx[i3,i2,i1+5]))

                    a2=y[4]*(Vy[i3,i2,i1]-Vy[i3,i2+1,i1])+\
                    y[3]*(Vy[i3,i2-1,i1]-Vy[i3,i2+2,i1])+y[2]*(Vy[i3,i2-2,i1]-Vy[i3,i2+3,i1])+\
                    y[1]*(Vy[i3,i2-3,i1]-Vy[i3,i2+4,i1])+y[0]*(Vy[i3,i2-4,i1]-Vy[i3,i2+5,i1])


                    a3=z[4]*(Vz[i3,i2,i1]-Vz[i3+1,i2,i1])+\
                    z[3]*(Vz[i3-1,i2,i1]-Vz[i3+2,i2,i1])+z[2]*(Vz[i3-2,i2,i1]-Vz[i3+3,i2,i1])+\
                    z[1]*(Vz[i3-3,i2,i1]-Vz[i3+4,i2,i1])+z[0]*(Vz[i3-4,i2,i1]-Vz[i3+5,i2,i1])
                    right=invD[i3,i2,i1]*(a1+a2+a3)
                    nex[i3,i2,i1]=2*cur[i3,i2,i1]-prev[i3,i2,i1]+dt2*right
```

We can vastly improve the time be rewriting the stepping function in numba.

```python
import numba
@numba.njit()

def oneStepNumba(Vx,Vy,Vz,invD,lamda,cur,prev,nex,dt,x,y,z):
        for i3 in range(5,Vx.shape[0]-5):
            for i2 in range(5,Vx.shape[1]-5):
                for i1 in range(5,Vx.shape[2]-5):
                    Vx[i3,i2,i1]=lamda[i3,i2,i1]*(\
                    x[4]*(cur[i3,i2,i1-1]-cur[i3,i2,i1])+\
                    x[3]*(cur[i3,i2,i1-2]-cur[i3,i2,i1+1])+\
                    x[2]*(cur[i3,i2,i1-3]-cur[i3,i2,i1+2])+\
                    x[1]*(cur[i3,i2,i1-4]-cur[i3,i2,i1+3])+\
                    x[0]*(cur[i3,i2,i1-5]-cur[i3,i2,i1+4])\
                                                 )

                    Vy[i3,i2,i1]=lamda[i3,i2,i1]*(y[4]*(cur[i3,i2-1,i1]-cur[i3,i2,i1])+\
                    y[3]*(cur[i3,i2-2,i1]-cur[i3,i2+1,i1])+y[2]*(cur[i3,i2-3,i1]-cur[i3,i2+2,i1])+\
                    y[1]*(cur[i3,i2-4,i1]-cur[i3,i2+3,i1])+y[0]*(cur[i3,i2-5,i1]-cur[i3,i2+4,i1]))    

                    Vz[i3,i2,i1]=lamda[i3,i2,i1]*(z[4]*(cur[i3-1,i2,i1]-cur[i3,i2,i1])+\
                    z[3]*(cur[i3-2,i2,i1]-cur[i3+1,i2,i1])+z[2]*(cur[i3-3,i2,i1]-cur[i3+2,i2,i1])+\
                    z[1]*(cur[i3-4,i2,i1]-cur[i3+3,i2,i1])+z[0]*(cur[i3-4,i2,i1]-cur[i3+4,i2,i1]))  


        dt2=dt*dt
        for i3 in range(10,nex.shape[0]-10):
            for i2 in range(10,nex.shape[1]-10):
                for i1 in range(10,nex.shape[2]-10):                   
                    a1=(x[4]*(Vx[i3,i2,i1]-Vx[i3,i2,i1+1])+\
                    x[3]*(Vx[i3,i2,i1-1]-Vx[i3,i2,i1+2])+x[2]*(Vx[i3,i2,i1-2]-Vx[i3,i2,i1+3])+\
                    x[1]*(Vx[i3,i2,i1-3]-Vx[i3,i2,i1+4])+x[0]*(Vx[i3,i2,i1-4]-Vx[i3,i2,i1+5]))

                    a2=y[4]*(Vy[i3,i2,i1]-Vy[i3,i2+1,i1])+\
                    y[3]*(Vy[i3,i2-1,i1]-Vy[i3,i2+2,i1])+y[2]*(Vy[i3,i2-2,i1]-Vy[i3,i2+3,i1])+\
                    y[1]*(Vy[i3,i2-3,i1]-Vy[i3,i2+4,i1])+y[0]*(Vy[i3,i2-4,i1]-Vy[i3,i2+5,i1])


                    a3=z[4]*(Vz[i3,i2,i1]-Vz[i3+1,i2,i1])+\
                    z[3]*(Vz[i3-1,i2,i1]-Vz[i3+2,i2,i1])+z[2]*(Vz[i3-2,i2,i1]-Vz[i3+3,i2,i1])+\
                    z[1]*(Vz[i3-3,i2,i1]-Vz[i3+4,i2,i1])+z[0]*(Vz[i3-4,i2,i1]-Vz[i3+5,i2,i1])
                    right=invD[i3,i2,i1]*(a1+a2+a3)
                    nex[i3,i2,i1]=2*cur[i3,i2,i1]-prev[i3,i2,i1]+dt2*right
```

We can then inherit a new class that overrides the step function with the Numba step function.


```python
class acousticDensityNumba(acousticDensity):
    def __init__(self,density,lamda,source,extract,dt):
        """
        density - Density as a function of space
        lamda - Lamda as a function of space
        nt    - Number of time samples
        source - Source function
        extract - Extract from wavefield
        
        """
        super().__init__(density,lamda,source,extract,dt)
    
    def stepField(self):
        oneStepNumba(self._Vx,self._Vy,self._Vz,self._invD._vals,self._lamda._vals,self._cur,self._prev,self._next,self._dt,self._x,self._y,self._z)
```

```python
@numba.njit(parallel=True)
def oalcVB(cur,Vx,Vy,Vz,lamda,x,y,z,b1,b2,b3,e1,e2,e3):
    for ib in numba.prange(len(b3)):
        for i3 in range(b3[ib],e3[ib]):
            for i2 in range(b2[ib],e2[ib]):
                for i1 in range(b1[ib],e1[ib]):
                   
                    Vx[i3,i2,i1]=lamda[i3,i2,i1]*(\
                    x[4]*(cur[i3,i2,i1-1]-cur[i3,i2,i1])+\
                    x[3]*(cur[i3,i2,i1-2]-cur[i3,i2,i1+1])+\
                    x[2]*(cur[i3,i2,i1-3]-cur[i3,i2,i1+2])+\
                    x[1]*(cur[i3,i2,i1-4]-cur[i3,i2,i1+3])+\
                    x[0]*(cur[i3,i2,i1-5]-cur[i3,i2,i1+4])\
                                                 )

                    Vy[i3,i2,i1]=lamda[i3,i2,i1]*(y[4]*(cur[i3,i2-1,i1]-cur[i3,i2,i1])+\
                    y[3]*(cur[i3,i2-2,i1]-cur[i3,i2+1,i1])+y[2]*(cur[i3,i2-3,i1]-cur[i3,i2+2,i1])+\
                    y[1]*(cur[i3,i2-4,i1]-cur[i3,i2+3,i1])+y[0]*(cur[i3,i2-5,i1]-cur[i3,i2+4,i1]))    

                    Vz[i3,i2,i1]=lamda[i3,i2,i1]*(z[4]*(cur[i3-1,i2,i1]-cur[i3,i2,i1])+\
                    z[3]*(cur[i3-2,i2,i1]-cur[i3+1,i2,i1])+z[2]*(cur[i3-3,i2,i1]-cur[i3+2,i2,i1])+\
                    z[1]*(cur[i3-4,i2,i1]-cur[i3+3,i2,i1])+z[0]*(cur[i3-4,i2,i1]-cur[i3+4,i2,i1]))  
@numba.njit(parallel=True)            
def oalcSB(Vx,Vy,Vz,invD,cur,prev,nex,dt2,x,y,z,b1,b2,b3,e1,e2,e3):
    for ib in numba.prange(len(b3)):
        for i3 in range(b3[ib],e3[ib]):
            for i2 in range(b2[ib],e2[ib]):
                for i1 in range(b1[ib],e1[ib]):
                    a1=(x[4]*(Vx[i3,i2,i1]-Vx[i3,i2,i1+1])+\
                    x[3]*(Vx[i3,i2,i1-1]-Vx[i3,i2,i1+2])+x[2]*(Vx[i3,i2,i1-2]-Vx[i3,i2,i1+3])+\
                    x[1]*(Vx[i3,i2,i1-3]-Vx[i3,i2,i1+4])+x[0]*(Vx[i3,i2,i1-4]-Vx[i3,i2,i1+5]))

                    a2=y[4]*(Vy[i3,i2,i1]-Vy[i3,i2+1,i1])+\
                    y[3]*(Vy[i3,i2-1,i1]-Vy[i3,i2+2,i1])+y[2]*(Vy[i3,i2-2,i1]-Vy[i3,i2+3,i1])+\
                    y[1]*(Vy[i3,i2-3,i1]-Vy[i3,i2+4,i1])+y[0]*(Vy[i3,i2-4,i1]-Vy[i3,i2+5,i1])


                    a3=z[4]*(Vz[i3,i2,i1]-Vz[i3+1,i2,i1])+\
                    z[3]*(Vz[i3-1,i2,i1]-Vz[i3+2,i2,i1])+z[2]*(Vz[i3-2,i2,i1]-Vz[i3+3,i2,i1])+\
                    z[1]*(Vz[i3-3,i2,i1]-Vz[i3+4,i2,i1])+z[0]*(Vz[i3-4,i2,i1]-Vz[i3+5,i2,i1])
                    
                    right=invD[i3,i2,i1]*(a1+a2+a3)
                    nex[i3,i2,i1]=2*cur[i3,i2,i1]-prev[i3,i2,i1]+dt2*right
```

```python
class blockClass:
    def __init__(self):
        pass;
    def calcBlocking(self,earth):
        raise Exception("Must override calcBlocking")
    def buildBlocks(self,b__1,b__2,b__3,e__1,e__2,e__3):
        b1=[]
        b2=[]
        b3=[]
        e1=[]
        e2=[]
        e3=[]
        for b_1 in b__1:
            for b_2 in b__2:
                for b_3 in b__3:
                    b1.append(b_1)
                    b2.append(b_2)
                    b3.append(b_3)
        for e_1 in e__1:
            for e_2 in e__2:
                for e_3 in e__3:
                    e1.append(e_1)
                    e2.append(e_2)
                    e3.append(e_3)
        return [np.asarray(b1),np.asarray(b2),np.asarray(b3)],\
                [np.asarray(e1),np.asarray(e2),np.asarray(e3)]
                    
class simpleBlock(blockClass):
    def __init__(self,nb):
        super().__init__()
        self._nb=nb
    
    def splitAxis(self,bs,n,edge):
        begA=[]
        endA=[]
        begA.append(edge)
        nleft=n-2*edge
        for i in range(bs):
            nlen=math.ceil(nleft/(bs-i))
            endA.append(begA[i]+nlen)
            if i!= bs-1:
                begA.append(endA[i])
            nleft-=nlen
        return begA,endA
    def calcBlocking(self,earth,edge):
        b1,e1=self.splitAxis(self._nb[0],earth.shape[0],edge)
        b2,e2=self.splitAxis(self._nb[1],earth.shape[1],edge)
        b3,e3=self.splitAxis(self._nb[2],earth.shape[2],edge)
        return self.buildBlocks(b1,b2,b3,e1,e2,e3)
```

```python
class acousticDensityBlock(acousticDensity):
    def __init__(self,density,lamda,source,extract,dt,blockC,stepF=None,sheerF=None,velocityF=None):
        """
        density - Density as a function of space
        lamda - Lamda as a function of space
        nt    - Number of time samples
        source - Source function
        extract - Extract from wavefield
        stepF - How to do a step (bloth velocity and sheer)
        blockC -  Blocking class
        sheerF,velocityF - Separate sheer and velocity functions
        """
        super().__init__(density,lamda,source,extract,dt)
        
        self._stepFunc=None
        if not isinstance(blockC,blockClass):
            raise Exception("Expecting block to be of a blocking class")
        if stepF==None:
            if velocityF ==None or sheerF==None:
                raise Exception("Must supply step or sheer and velocity")
        
        self._sheerFunc=sheerF
        self._velFunc=velocityF
        self._stepFunc=stepF
        self._blockC=blockClass
        
        self._bV,self._eV=blockC.calcBlocking(density._vals,5)
        self._bS,self._eS=blockC.calcBlocking(density._vals,10)
    def stepField(self):
        if self._stepFunc!=None:
            self._stepFunc(self._Vx,self._Vy,self._Vz,self._invD._vals,self._lamda._vals,\
                           self._cur,self._prev,self._next,self._dt,self._x,self._y,self._z,\
                           self._bV[0],self._bV[1],self._bV[2],self._eV[0],self._eV[1],\
                           self._eV[2])
        else:
            self._velFunc(self._cur,self._Vx,self._Vy,self._Vz,self._invD._vals,\
                          self._x,self._y,self._z,   \
                          self._bV[0],self._bV[1],self._bV[2],self._eV[0],self._eV[1],\
                           self._eV[2])
            self._sheerFunc(self._Vx,self._Vy,self._Vz,self._invD._vals,\
                          self._cur,self._prev, self._next,self._dt*self._dt,\
                            self._x,self._y,self._z,\
                          self._bV[0],self._bV[1],self._bV[2],self._eV[0],self._eV[1],\
                           self._eV[2])
```

```python
import SepVector
import pyAcousticDensity
class acousticDensityCpp:
    def __init__(self,density,lamda,source,islice,dt,blockC):
        """
        density - Density as a function of space
        lamda - Lamda as a function of space
        nt    - Number of time samples
        source - Source function
        islice - Slice to store
        blockC -  Blocking class
        """
        self.baseInit(density,lamda,source,islice,dt,blockC)
        self._cppMode=pyAcousticDensity.acousticDensity(self._density.cppMode,self._lamda.cppMode,\
         self._source._ix, self._source._iy,self._source._iz,self._sval.cppMode,\
        self._source._maxF,self._dt,self._islice,self._bV,self._eV,self._bS,self._eS)
        
    def baseInit(self,density,lamda,source,islice,dt,blockC):
        
        if not isinstance(blockC,blockClass):
            raise Exception("Expecting block to be of a blocking class")
            
        b_V,e_V=blockC.calcBlocking(density.getNdArray(),5)
        b_S,e_S=blockC.calcBlocking(density.getNdArray(),10)
        self._density=density
        self._lamda=lamda
        self._source=source
        self._bS=[]
        self._bV=[]
        self._eS=[]
        self._eV=[]
        self._dt=dt
        self._islice=islice
        for b in b_V:
            self._bV.append(b.tolist())
        for b in b_S:
            self._bS.append(b.tolist())
        for e in e_V:
            self._eV.append(e.tolist())
        for e in e_S:
            self._eS.append(e.tolist())
        self._sval=SepVector.getSepVector(source._vals.astype(np.float32))
    
        
    def prop(self,nt):
        self._cppMode.prop(nt)
    def getWavefield(self):
        return SepVector.floatVector(fromCpp=self._cppMode.getWavefield())
```

```python
class acousticDensityISPC(acousticDensityCpp):
    def __init__(self,density,lamda,source,islice,dt,blockC):
        """
        density - Density as a function of space
        lamda - Lamda as a function of space
        nt    - Number of time samples
        source - Source function
        islice - Slice to store
        blockC -  Blocking class
        """
        self.baseInit(density,lamda,source,islice,dt,blockC)
        self._cppMode=pyAcousticDensity.acousticDensityISPC(self._density.cppMode,self._lamda.cppMode,\
         self._source._ix, self._source._iy,self._source._iz,self._sval.cppMode,\
        self._source._maxF,self._dt,self._islice,self._bV,self._eV,self._bS,self._eS)
wave=acousticDensityISPC(dens,lamda,mySource,100,.002,blockC)
       
```
